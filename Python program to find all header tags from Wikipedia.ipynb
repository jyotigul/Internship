```python
from bs4 import BeautifulSoup
import requests
```

```python
page =requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
page
```

```python
soup =BeautifulSoup(page.content)
soup
```

```python
title= []
for i in soup.find_all(['h1','h2','h3','h4','h5','h6']):
    title.append(i.text.replace('\n','').replace('\xa0...',''))
    
title
```

```python
import pandas as pd
df=pd.DataFrame({'Header':title})
df
```
