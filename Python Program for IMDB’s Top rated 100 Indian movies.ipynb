```python
from bs4 import BeautifulSoup
import requests
```

```python
page =requests.getpage =requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
name =[]
for i in soup.find_all('td',class_="titleColumn"):
    name.append(i.find('a').text)
    
name
```

```python
time =[]
for i in soup.find_all('td',class_="titleColumn"):
    time.append(i.find('span',class_="secondaryInfo").text.replace('(','').replace(')','').replace(' ',''))
    
time
```

```python
rating =[]
for i in soup.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text.replace('\n',''))
    
rating
```

```python
print(len(name),len(time),len(rating))
```

```python
import pandas as pd
df = pd.DataFrame({})
df['Movie Name']=name[:100]
df['Rating']=rating[:100]
df['Year of Release']=time[:100]
df
```
