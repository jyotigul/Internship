```python
from bs4 import BeautifulSoup
import requests
```

```python
page =requests.getpage =requests.get('https://presidentofindia.nic.in/former-presidents.htm')
```

```python
page
```

```python
soup =BeautifulSoup(page.content)
soup
```

```python
presidents = []

for president_name in soup.find_all('div',class_="presidentListing"):
    presidents.append(president_name.h3.text)
    
presidents
```

```python
terms = []

for president_term in soup.find_all('div',class_="presidentListing"):
    terms.append(president_term.p.text.replace('Term of Office:',''))
    
terms
    
```

```python
print(len(presidents),len(terms))
```

```python
import pandas as pd
df = pd.DataFrame({'Name':presidents,'Term of Office':terms})
df
```
